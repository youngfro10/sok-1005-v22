```python
# Packages
# Import for statsmodels
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from statsmodels.regression.linear_model import OLS

# Bokeh pakker
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.plotting import output_file
# Disable chained assignments
pd.options.mode.chained_assignment = None 
```

```python
#creating a dictionary of continents with norwegian translation:
Countries={'Brazil':'Brasil','Russia':'Russland','Nigeria':'Nigeria','Japan':'Japan', 'USA':'USA', 'Spain':'Spania', 'Italy':'Italia', 'Norway':'Norge', 'UK':'Storbritannia', 'Denmark':'Danmark', 'Sweden':'Sverige', 'Greenland':'Grønnland','India':'India','France':'Frankrike','China':'Kina', 'Germany':'Tyskland','Canada':'Kanada'}
#assigning a color hex-code to each continent
colors={'Brazil':'#E14823','Russia':'#050505', 'Nigeria':'#E14833','Japan': '#E14827', 'USA': '#84E127', 'Spain': '#2792E1', 'Italy': '#E04A6C', 'Norway' : '#DEB887', 'UK': 'darkgreen', 'Denmark': 'purple', 'Sweden': '#5F9EA0', 'India': '#458B74', 'Greenland':'#E3CF57', 'France':'#FF6103', 'China':'#458B00', 'Germany':'#6495ED', 'Canada':'#BDB76B'}


gg['Land']=gg['Countries'].apply(lambda cou:  Countries[cou])
gg['colors']=gg['Countries'].apply(lambda cou: colors[cou])
gg
```

```python
gg['Populasjon'] = gg['Population'].apply(lambda pop: pop**0.5/300)
gg
```

```python
y=gg['TotalDeaths']
pd.DataFrame(y)
```

```python
x=pd.DataFrame((gg['Population']))
x['intercept']=1
x
```

```python
res=OLS(y,x).fit()

print(res.summary())
```

```python
x=np.linspace(min(np.log(gg['Population'])), max(np.log(gg['Population'])), 100)

regression_line=res.params['intercept']+res.params['Population']*x

ax.plot(x, regression_line,color='red')
fig
```

```python
import seaborn as sns
sns.regplot(x='Population', y= 'TotalDeaths', data=gg)
```

```python
gg['Populasjon']=gg['Population'].apply(lambda pop:  '{:,}'.format(int(pop/1000000)))
gg['Dødsfall']=gg['TotalDeaths'].apply(lambda l: int(l))
gg
```

```python
# Creating figure:
p = figure(
        title = "Total Dødsfall", 
        x_axis_label = 'Populasjon',
        y_axis_label = 'DeathCases',
    
        x_axis_type = "log",
        y_axis_type = "log",
        tools="hover", 
        tooltips = [
            ("Land", "@Land"),
            ("Dodsfall", "@TotalDeaths"),
            ("Befolkning", "@Populasjon"),
            ],
    
        plot_height = 1000,
        plot_width = 1700)
```

```python
p.scatter(
    source=gg,
    x= 'Populasjon', 
    y= 'TotalDeaths',
    
    size='Populasjon', 
    color= 'colors',
    legend_field='Land',
    alpha= 0.8
    )

# Formatting:
p.xaxis.major_label_overrides = { 1000: '1k', 10000: '10k', 100000: '100k', 1000000: '1m' }
p.legend.location = "top_left"

p.legend[0].border_line_alpha=0
p.outline_line_alpha=0
p.grid[0].grid_line_alpha=0
p.grid[1].grid_line_alpha=0

# Creating the graph and saving as html
output_notebook()
output_file("Dødsfall.html")
show(p)
```
